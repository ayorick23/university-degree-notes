```python
import pandas as pd
```

## 1. Leer el archivo empleados.csv

```python
df = pd.read_csv('empleados.csv', encoding="latin-1") #UTF-8 no lo soportó
```

```python
#Convertir a un DataFrame
data = pd.DataFrame(df)
```

```python
print(data)
```

## 2. Mostrar el promedio de salario por departamento

```python
salario_prom_departamento = data.groupby('Departamento')['Salario'].mean()
print(salario_prom_departamento)
```

## 3. Mostrar el empleado de mayor edad

```python
print(data['Edad'].max())
```

```python
print(data[data['Edad'] == data['Edad'].max()])
```

## 4. Calcular cuántos empleados hay en cada departamento

```python
empleados_departamento = data.groupby('Departamento')['Nombre'].count()
print(empleados_departamento)
```

## 5. Exportar resumen_departamentos.csv

```python
resumen = {
    'Departamento': empleados_departamento.index,
    'Total Empleados': empleados_departamento.values,
    'Salario Promedio': salario_prom_departamento.values
}
```

```python
df_resumen = pd.DataFrame(resumen)
print(df_resumen)
```

```python
df_resumen.to_csv('resumen_departamentos.csv', index=False)
```
